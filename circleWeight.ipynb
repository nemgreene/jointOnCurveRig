```python
import matplotlib.pyplot as plt
import numpy as np
import math

```

```python
xList = list(range(11))

def faloff(x, cvs, offset):
    b = abs((len(cvs) -1 )/2 - x)
    c = (len(cvs) - 1 )/2
    a = math.sqrt(abs(c**2-b **2))
    return round(a/c * offset -.5, 4) 
y = list(map(lambda x : faloff(x, xList, .5), xList))
z = list(map(lambda x : faloff(x, xList, -.5), xList))

fig, ax = plt.subplots()

for i, txt in enumerate(y):
    ax.annotate(txt, (xList[i], y[i]))

for i, txt in enumerate(z):
    ax.annotate(txt, (xList[i], z[i]))
plt.plot(xList, y)
plt.plot(xList, z)

```
